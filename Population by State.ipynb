This notebook takes census data from the web and converts it into a pandas dataframe that we will merge with another dataframe on school shotting data.  We wanted to have this data to compair the number of shootings per capita by state to gun control laws to see if there is a corrilation between stronger gun control and less school shootings, or if can just be explained by the population of each state.

```python
import pandas as pd
```

```python
df1 = pd.read_excel('nst-est2017-01.xlsx')
#the data came from the folowing website: which no longer exsists so we had to use web archive to find it
#https://www.census.gov/data/datasets/2017/demo/popest/state-total.html
#https://web.archive.org/web/20180820005818/https://www2.census.gov/programs-surveys/popest/tables/2010-2017/state/totals/nst-est2017-01.xlsx
```

```python
df1.head()
```

```python
# change the names of the columns to account for the extra rows at the top of the excel file
df1.columns = df1.iloc[2]
```

```python
df1.head()
```

```python
#drop the row from the top that has no data
df1 = df1.drop([0])
```

```python
df1.head()
```

```python
#drop the row from the top that has no data
df1 = df1.drop([1])
```

```python
df1.head()
```

```python
#drop the rows with the regional data
df1 = df1.drop([2,4,5,6,7])
```

```python
df1.head()
```

```python
#drop the rows with data not from states
df1 = df1.drop([59,60,61,62,63,64,65])
```

```python
df1.head()
```

```python
# rename columns to get rid of extra white space and decimals
df1 = df1.rename(columns={            2011.0: 2011,           2012.0: 2012,           2013.0:2013,           2014.0: 2014,           2015.0:2015,           2016.0:2016,           2017.0:2017})
```

```python
df1.columns
```

```python
df1.iloc[:, 0]
```

```python
# Add new column with States names without the period
df1["State"] = [' United States','Alabama','Alaska','Arizona','Arkansas','California','Colorado','Connecticut','Delaware','District of Columbia','Florida','Georgia','Hawaii','Idaho','Illinois','Indiana','Iowa','Kansas','Kentucky','Louisiana','Maine','Maryland','Massachusetts','Michigan','Minnesota','Mississippi','Missouri','Montana','Nebraska','Nevada','New Hampshire','New Jersey','New Mexico','New York','North Carolina','North Dakota','Ohio','Oklahoma','Oregon','Pennsylvania','Rhode Island','South Carolina','South Dakota','Tennessee','Texas','Utah','Vermont','Virginia','Washington','West Virginia','Wisconsin','Wyoming']
```

```python
df1.head()
```

```python
#drop the states with the periods
df = df1.drop(df1.columns[0], axis=1)
```

```python
#drop unnessisary columns
df = df.drop(columns=['Census', 'Estimates Base'])
```

```python
df.head()
```

```python
df2018 = pd.read_csv('Population Estimates by State.csv')
#this data is only the 2018 population by state from the census
```

```python
df2018.head()
```

```python
#rename column for easier joining in the future
df2018 = df2018.rename(columns={'Population Estimates by State': "State"})
```

```python
df2018.head()
```

```python
df2018.columns = df2018.iloc[0]
```

```python
df2018.head()
```

```python
#rename columns for easier joining in the future
df2018 = df2018.rename(columns={'States': "State", 'Population':2018})
```

```python
df2018.head()
```

```python
#merege the 2018 data to the rest of the data
result = pd.merge(df, df2018, how='left', on=['State'])
```

```python
result.head()
```

```python
#get rid of the floats and turn data into ints
result[2011] = result[2011].apply(lambda x: int(x))
```

```python
result.head()
```

```python
#yes, I know this isn't DRY code. So sue me, I was being lazy
result[2012] = result[2012].apply(lambda x: int(x))
result[2013] = result[2013].apply(lambda x: int(x))
result[2014] = result[2014].apply(lambda x: int(x))
result[2015] = result[2015].apply(lambda x: int(x))
result[2016] = result[2016].apply(lambda x: int(x))
result[2017] = result[2017].apply(lambda x: int(x))
```

```python
result.head()
```

```python
#our other data uses state abbrivations instead of the full names, so in order to join that data we added in the state codes
result["Code"] = ['US','AL','AK','AZ','AR','CA','CO','CT','DE','DC','FL','GA','HI','ID','IL','IN','IA','KS','KY','LA','ME','MD','MA','MI','MN','MS','MO','MT','NE','NV','NH','NJ','NM','NY','NC','ND','OH','OK','OR','PA','RI','SC','SD','TN','TX','UT','VT','VA','WA','WV','WI','WY']
```

```python
result
```

```python
#no longer need the full state names so let's drop them
result = result.drop(columns=['State'])
```

```python
result
```

```python
#we don't need the overall US population so we drop it
result = result.drop([0])
```

```python
result
```

```python
#changing the name to make the join easier
result = result.rename(columns={'Code': "State"})
```

```python
result.head()
```

```python
#save to a csv to merge with other data
result.to_csv('population.csv')
```
